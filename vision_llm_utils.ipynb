```python
import cv2
import os
import matplotlib.pyplot as plt


def  preprocess_for_vision_llm(input_path,output_path,max_width=640,max_height=640):
  """
  Resize and save an image optimized for Vision LLM token efficiency.
  Maintain aspect ratio within max dimensions.

  """

  image = cv2.imread(input_path)

  height, width = image.shape[:2]
  scale = min(max_width / width, max_height / height)
  new_w, new_h = int(width*scale),int(height*scale)

  resized = cv2.resize(image,(new_w,new_h))
  cv2.imwrite(output_path,resized)
  print(f"Saved: {new_w}x{new_h} -> {output_path}")


# Example usage
preprocess_for_vision_llm("star-bunny.jpeg","resized-bunny.jpeg")
```
