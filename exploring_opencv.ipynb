```python
import matplotlib.pyplot as plt
import cv2

# Load an image
image = cv2.imread('star-bunny.jpeg')

#display the original and resized images
plt.imshow(image)
plt.title('resized image')
plt.show()
```

```python
resized = cv2.resize(image, (150,150))

# display the resized star bunny
plt.imshow(resized)
plt.title('resized bunny')
plt.show()
```

```python
cv2.destroyAllWindows()
```

```python
# cropping the image
cropped_image = image[0:100,0:100]

# show the cropped image
plt.imshow(cropped_image)
plt.title('cropped star bunny')
plt.show()
```

```python
center = (image.shape[0] // 2,image.shape[1] // 2)
matrix = cv2.getRotationMatrix2D(center,45,1) # 90 degree rotation
rotated = cv2.warpAffine(image,matrix,(image.shape[0],image.shape[1]))

# show the rotated image
plt.imshow(rotated)
plt.title('rotated star bunny')
plt.show()
```
